### Grouping

#### simple example

```python
names = ['John', 'Matt', 'Sara', 'Jim', 'Ashley']
ids = [ 23, 34, 83, 86, 12]
balance = [10.2, 84.3, 72.9, 27.1, 223.1]
department = ['A','A','B','B','B']

users = pd.DataFrame({'id': ids, 
                      'name': names, 
                      'bal':balance, 
                      'dept': department})
print users.head()
```

```python
users = users.reindex(columns=['name','id','dept','bal'])
print users
```

```python
# what's the total balance per dept?
dept = users[['dept','bal']]
print dept
```

```python
grps = dept.groupby('dept')
```

```python
grps.groups
```

```python
print 'number of groups:',len(grps)
print type(grps)
print type(grps.groups)
print grps.groups
```

```python
for key, value in grps:
    print key, type(value)
    print value
    print ''
```

```python
# apply an aggregate function to the groups, in this example - sum
print grps.sum()
```

```python
print dept.groupby('dept').sum()
```
